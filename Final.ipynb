```python
import pandas as pd
import altair as alt
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path="C:\\Users\pchandra\Desktop\employees.csv"
df=pd.read_csv(file_path)
df=df.dropna()
print(df)
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x="Team",
    y="count():Q")
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department"),
    y="count():Q")
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    )
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    )
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    ).properties(title="Employees per Department")
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    color="Team"
    ).properties(title="Employees per Department")
bar_graph.display()
```

```python
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    color=alt.Color("Gender:N", legend=alt.Legend(title="Gender")
    )).properties(title="Employees per Department")
bar_graph.display()
```

```python
interval = alt.selection_interval()
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    color=alt.Color("Gender:N", legend=alt.Legend(title="Gender")
    )).add_selection(interval)
bar_graph.display()
```

```python
sel = alt.selection(type="single",fields=["Gender"])
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    color=alt.Color("Gender:N", legend=alt.Legend(title="Gender")),
    opacity=alt.condition(sel,alt.value(1),alt.value(0.2))
    ).add_selection(sel)
bar_graph.display()
```

```python
point_nearest = alt.selection_point(on='pointerover', nearest=True)
bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count():Q",title="No. of employees"),
    color=alt.Color("Gender:N", legend=alt.Legend(title="Gender")),
    opacity=alt.condition(point_nearest,alt.value(1),alt.value(0.2)),
    tooltip=["Team","Gender"]
    ).add_params(point_nearest)
bar_graph.display()
```

```python
point_nearest = alt.selection_point(on='pointerover', nearest=True)
selection = alt.selection_single(fields=['Gender'])
color = alt.condition(selection,alt.Color('Gender:N').legend(None),alt.value('lightgray'))

bar_graph=alt.Chart(df).mark_bar().encode(
    x=alt.X("Team:N",title="Department",axis=alt.Axis(labelAngle=-45)),
    y=alt.Y("count:Q",title="No. of employees"),
    color=color,
    opacity=alt.condition(point_nearest,alt.value(1),alt.value(0.2)),
    tooltip=["Team","count:Q","Gender"]
    ).transform_aggregate(groupby=["Team","Gender"],count="count()").add_params(point_nearest,selection)

legend = alt.Chart(df).mark_point().encode(
    alt.Y('Gender:N').axis(orient='right'),
    color=color
).add_params(
    selection
)

bar_graph|legend
```
