```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import altair as alt
```

```python
file_path="C:\\Users\pchandra\Desktop\employees.csv"
df=pd.read_csv(file_path)
df=df.dropna()
print(df)
```

```python
df_=df[(df["Team"]=="Business Development")|(df["Team"]=="Distribution")|(df["Team"]=="Legal")]
print(df_)
```

```python
c=alt.Chart(df_).mark_arc(innerRadius=0).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Team"),
    tooltip=alt.Tooltip("Team")
)
c
```

```python
c_Team=alt.Chart(df_).mark_arc(innerRadius=0).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Team"),
    tooltip=alt.Tooltip("Team")
)
c_Gender=alt.Chart(df_).mark_arc(innerRadius=0).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Gender"),
    tooltip=alt.Tooltip("Gender"))
c_Team|c_Gender
```

```python
c_Team=alt.Chart(df_).mark_arc(innerRadius=100).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Team"),
    tooltip=alt.Tooltip("Team")
)

c_Gender=alt.Chart(df_).mark_arc(innerRadius=50).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Gender"),
    radius=alt.Radius(value=100),
    tooltip=alt.Tooltip("Gender"))
c_Team+c_Gender
```

```python
c_Team=alt.Chart(df_).mark_arc(innerRadius=100).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Team"),
    tooltip=alt.Tooltip("Team")
)

c_Gender=alt.Chart(df_).mark_arc(innerRadius=50).encode(
    theta=alt.Theta("count()",title="No. of employees in a department"),
    color=alt.Color("Gender"),
    radius=alt.Radius(value=100),
    tooltip=[alt.Tooltip("Gender:N"), alt.Tooltip("count:Q")]).transform_aggregate(
    count="count(*)",groupby=["Gender"])
c_Team+c_Gender
```
